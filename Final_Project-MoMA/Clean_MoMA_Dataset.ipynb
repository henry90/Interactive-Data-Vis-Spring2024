```python
import nltk
from nltk import pos_tag
from nltk.chunk import ne_chunk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet as wn
import pycountry
import pandas as pd
import re
```

```python
data = pd.read_csv('Artworks.csv')
data.head(10)
```

```python
import re
import numpy as np

# Extract years from a date string
def extract_years(date_str):
    # Find all four-digit numbers (years) in the string
    year_matches = re.findall(r'\d{4}', str(date_str))
    return list(map(int, year_matches))  # Convert to integers

data['extracted_years'] = data['Date'].apply(extract_years)

# Filter rows with no data
data = data[data['extracted_years'].apply(lambda x: len(x) > 0)]

# Get the start year (the earliest year in the list)
data['start_year'] = data['extracted_years'].apply(lambda x: min(x) if x else np.nan)

# Get the end year (the latest year in the list)
data['end_year'] = data['extracted_years'].apply(lambda x: max(x) if x else np.nan)

# Convert to integers
data['start_year'] = data['start_year'].astype(int)
data['end_year'] = data['end_year'].astype(int)

data
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

nltk.download('punkt')  # Tokenizer
nltk.download('averaged_perceptron_tagger')  # POS tagger

# Extract noun phrases as potential materials
def extract_nouns(text):
    # Tokenize the text
    tokens = word_tokenize(str(text))
    # Part-of-speech tagging
    pos_tags = pos_tag(tokens)
    # Extract nouns and filter out unwanted words
    unwanted_words = {'Artist'}
    nouns = [word for word, tag in pos_tags if tag.startswith('NN') and word not in unwanted_words]  # NN: Singular noun
    return list(set(nouns))  # Return unique nouns

data['extracted_materials'] = data['Medium'].apply(extract_nouns)

data
```

```python
import ast

def safe_literal_eval(val):
    try:
        if isinstance(val, list):
            return val
        elif isinstance(val, str) and val.startswith('[') and val.endswith(']'):
            return ast.literal_eval(val)
        else:
            return [int(val)] if val else []
    except (ValueError, SyntaxError):
        return []

# Convert string representations of lists into actual lists
data['extracted_years'] = data['extracted_years'].apply(safe_literal_eval)
data['extracted_materials'] = data['extracted_materials'].apply(safe_literal_eval)

# Delete rows with missing values in 'extracted_years', 'extracted_materials, 'start_year', and 'end_year'
data = data.dropna(subset=['extracted_years', 'extracted_materials', 'start_year', 'end_year'])

# Fill missing values in 'start_year' and 'end_year'
data['start_year'] = data['start_year'].fillna(data['extracted_years'].apply(lambda x: x[0] if x else None))
data['end_year'] = data['end_year'].fillna(data['extracted_years'].apply(lambda x: x[-1] if x else None))

print(data.head())
```

```python
from collections import Counter
import numpy as np
import json

# Determine the decade from a year
def get_decade(year):
    return (year // 10) * 10

decade_materials = []

for index, row in data.iterrows():
    start_year = row['start_year']
    end_year = row['end_year']
    materials = row['extracted_materials']

    if pd.notnull(start_year) and pd.notnull(end_year) and materials:
        start_decade = get_decade(int(start_year))
        end_decade = get_decade(int(end_year))
        for year in range(start_decade, end_decade + 1, 10):
            decade_materials.append((year, row['Department'], materials))

# Create a dataframe from the results
decade_materials_df = pd.DataFrame(decade_materials, columns=['Decade', 'Department', 'Materials'])

# Explode the materials list into separate rows
decade_materials_df = decade_materials_df.explode('Materials')

# Group by Decade, Department, and Materials to count occurrences
material_counts = decade_materials_df.groupby(['Decade', 'Department', 'Materials']).size().reset_index(name='Count')

# Get the top 10 materials for each decade and department
top_materials = material_counts.groupby(['Decade', 'Department']).apply(lambda x: x.nlargest(10, 'Count')).reset_index(drop=True)

print(top_materials.head())
```

```python
# Prepare the data in a nested dictionary format
visualization_data = {}

for decade, group in top_materials.groupby('Decade'):
    visualization_data[decade] = {}
    for department, dept_group in group.groupby('Department'):
        visualization_data[decade][department] = dept_group[['Materials', 'Count']].to_dict(orient='records')

# Save JSON file
with open('top_10_materials.json', 'w') as json_file:
    json.dump(visualization_data, json_file, indent=4)
```
